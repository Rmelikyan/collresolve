```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
with open('c2019_1myr_colllog.txt', 'r') as f:
    lines = f.readlines()
```

```python
num_ps_init = 156
num_ps = num_ps_init
num_ps_array = []
num_hnr = 0 
times = []
for line in lines:
    if 'collision' in line:
        times.append(float(line.split(' ')[3]))
        num_ps -= 1
    if 'Body' in line:
        second_body_mass = float(line.split(',')[8].split('=')[-1])
        if second_body_mass  > 0:
            num_ps += 1
            num_hnr += 1
        if second_body_mass < 0:
            print("That's interesting")
        num_ps_array.append(num_ps)
print('Total number of Collisions {}; Number of HitNRuns {}'.format(len(times), num_hnr))
```

```python
plt.plot(np.sort(times), range(len(times)))
plt.plot(np.sort(times), num_ps_array)
fig = plt.gcf(); ax = plt.gca()
num_ps
```
